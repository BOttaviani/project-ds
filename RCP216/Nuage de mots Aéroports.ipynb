```python
import pandas, os, json
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline

df = pandas.read_csv('centralite.csv')
df.dtypes
#transformation en dictionnaire
#data=df[df["centralite"]>10000].set_index('aeroport')['centralite'].to_dict()
data=df.set_index('aeroport')['centralite'].to_dict()
data
```

```python
wc = WordCloud(font_path="/usr/share/fonts/truetype/lato/Lato-Heavy.ttf",max_words=300,
               width=1200,height=800, 
               background_color='grey',#relative_scaling=.9, 
                  colormap="plasma",prefer_horizontal=0.8).generate_from_frequencies(data)
%matplotlib inline
plt.figure(figsize=(20,15))
plt.title('Aéroports ayant la plus forte centralité', fontsize=30, color='blue')
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python

wc = WordCloud(font_path="/usr/share/fonts/truetype/tlwg/TlwgMono-Bold.ttf",max_words=100,
               width=1200,height=800, 
               background_color='white',#relative_scaling=.9, 
                  colormap="viridis",prefer_horizontal=0.8).generate_from_frequencies(data)
%matplotlib inline
plt.figure(figsize=(20,15))
plt.title('Aéroports ayant la plus forte centralité', fontsize=30)
plt.imshow(wc)
plt.axis("off")
plt.show()
```
